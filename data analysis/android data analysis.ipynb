```python
import pandas as pd
import numpy as np
df_test=pd.read_excel(r"C:\Users\Mahmood\OneDrive\Desktop\datafinallllll.xlsx")
df_test
```

```python
df_test=df_test[["Model","RAM","Storage","Condition","FIX","Price"]]
df_test.head(2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. تحديد العواميد الرقمية اللي هنشوف الارتباط بتاعها
numeric_cols = ['Storage', 'RAM', 'Condition', 'FIX', 'Price']

# 2. حساب مصفوفة الارتباط (Correlation Matrix)
corr_matrix = df_test[numeric_cols].corr()

# 3. رسم الـ Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn', fmt=".2f", linewidths=0.5)

plt.title('Correlation between Features and Price')
plt.show()

# 4. ترتيب الارتباط مع السعر تنازلياً عشان تشوف الأهم فالأهم
print("--- ترتيب قوة الارتباط مع السعر ---")
print(corr_matrix['Price'].sort_values(ascending=False))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# رسمة توضح توزيع السعر بناءً على الحالة (Condition)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Condition', y='Price', data=df_test, palette='viridis')

plt.title('Price Distribution by Condition')
plt.xlabel('Condition (1: Poor, 4: Excellent)')
plt.ylabel('Price (EGP)')
plt.show()
```
