```python
import pandas as pd
import numpy as np
df_test=pd.read_excel(r"C:\Users\Mahmood\Downloads\data\realme.      .xlsx")
df_test
```

```python
df_test=df_test[["BRAND","Model","RAM","Storage","Condition","FIX","Price"]]
df_test.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
# تحويل كل القيم في عمود Model ل string
df_test['Model'] = df_test['Model'].astype(str)

#هنعمل دلوقتي LabelEncoder 
le_model = LabelEncoder()
df_test['Model_encoded'] = le_model.fit_transform(df_test['Model'])


print(df_test[['Model', 'Model_encoded']].head())
```

```python
# إنشاء الديكشنري النهائي للـ GUI
brand_model_mapping = {}

for index, row in df_test.iterrows():
    # دمج البراند مع الموديل (مثلاً: Samsung + S23 Ultra)
    full_name = f"{row['BRAND']} {row['Model']}"
    
    # إضافة الاسم كـ Key والكود كـ Value
    brand_model_mapping[full_name] = row['Model_encoded']

# عرض أول 10 عناصر للتأكد
print("النتائج النهائية للديكشنري (الاسم الكامل : الكود):")
for name, code in list(brand_model_mapping.items())[:10]:
    print(f"{name} -> {code}")
```
